```python
# Моделирование магнитной восприимчивости для теннантита Cu--Sb--Se
sus_exp_Cu_Sb_Se = pd.read_csv('!!!Cu3SbSe3.txt', '\t', names = ['temperature','mag_sus']) # импорт экспериментальных данных

# Моделирование теплоемкости для теннантита sus_exp_Cu_Sb_Se
#fig = pp.figure()
#fig.clear()



#pp.plot(sus_exp_Cu_Sb_Se['temperature'],sus_exp_Cu_Sb_Se['mag_sus'], 'co')
#pp.xlabel(u'Температура, К')    # обозначение оси абсцисс
#pp.ylabel(u'Магнитная восприимчивость')    # обозначение оси ординат
#pp.rc('axes', linewidth=1) 

#pp.savefig('sus_exp_Cu_Sb_Se.png', dpi = 600)

#pp.show()

x = sus_exp_Cu_Sb_Se['temperature'][270:298]
y = sus_exp_Cu_Sb_Se['mag_sus'][270:298]

def func( x, a, b):
    return a + b/(x)
 
p0 = sy.array( [1, 1, 1] )
coeffs, matcov = curve_fit( func, x, y )
#yaj = func( x, coeffs[0], coeffs[1], coeffs[2] )

full_price = (sus_exp_Cu_Sb_Se.mag_sus - coeffs[0])
sus_exp_Cu_Sb_Se['mag_sus_without_dia'] = full_price

fig = pp.figure()
fig.clear()



pp.plot(sus_exp_Cu_Sb_Se['temperature'],sus_exp_Cu_Sb_Se['mag_sus_without_dia']*10**8, 'co')
pp.xlabel(u'Температура, К', fontsize=16)    # обозначение оси абсцисс
pp.ylabel(r'Восприимчивость, $10^{-8}$', fontsize=18)     # обозначение оси ординат
plt.tight_layout()  
#pp.rc('axes', linewidth=1) 
pp.ylim((0.4,0.8))
pp.xlim((0,280))
pp.tick_params(axis='both', labelsize = 15)
pp.rc('axes', linewidth=3)

pp.tight_layout()
pp.savefig('sus_exp_Cu_Sb_Se.png', dpi = 600)

pp.show()
print coeffs[0]
```

```python
# Моделирование магнитной восприимчивости для теннантита Cu--Sb--S
sus_exp_Cu_Sb_S = pd.read_csv('!!Cu3SbS3_2.txt', '\t', names = ['temperature','mag_sus']) # импорт экспериментальных данных



x = sus_exp_Cu_Sb_S['temperature']
y = (sus_exp_Cu_Sb_S['mag_sus']) 

def func( x, a, b, c):
    return a + b / (x + c)
value_left, value_right = 0,70 

X_for_dia = x[value_left:value_right]
coeffs, matcov = curve_fit( func
                           , X_for_dia
                           , y[value_left:value_right]
#                            p0 = p0, 
#                            ,method = 'lm',  
#                            bounds=([0.1, 40], [1., 100]) 
                          )
yaj = func( X_for_dia
           , coeffs[0]
           , coeffs[1]
           , coeffs[2]
          )


fig = pp.figure()
fig.clear()



pp.plot(x, (y - coeffs[0])*10**4, 'co'
        , X_for_dia, (yaj - coeffs[0])*10**4,'r-'
       )
pp.xlabel(u'Температура, К', fontsize=16)    # обозначение оси абсцисс
pp.ylabel(r'Восприимчивость$\times10^{-4}$', fontsize=18)      # обозначение оси ординат

pp.rc('axes', linewidth=1) 
pp.ylim((0.1,2))
pp.xlim((-2,300))

pp.tick_params(axis='both', labelsize = 15)
pp.rc('axes', linewidth=3)

pp.tight_layout()

pp.text(200, 1.7
        , r'$\Theta=$0.25 K'
        , fontsize=18
       )
pp.savefig('sus_exp_Cu_Sb_S.png', dpi = 600)


pp.show()
y_dia = coeffs[0]
y_cur = coeffs[1]
coeffs
```
