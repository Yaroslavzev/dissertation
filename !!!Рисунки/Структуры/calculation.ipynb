```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#%matplotlib qt5 #interactive plotting in separate window %matplotlib qt  
#%matplotlib notebook #interactive charts inside notebooks, matplotlib 1.4+

%matplotlib inline 
```

```python
calculation = pd.read_csv('calculation.txt', '\t')
                       #, names = ['temperature','heat_capacity']) # импорт экспериментальных данных

list(calculation)
print calculation
```

```python
fig = plt.figure()
fig.clear()

plt.scatter(calculation['\xb9_structure'],calculation['energy'])


plt.xlabel(u'Номер структуры', fontsize=16)
plt.ylabel(u'Энергия эл. ячейки, эВ', fontsize=16)

A = 0, 19
B = -246.13, -246.13
plt.plot(A,B, 'r--', linewidth=3.0)
#plt.text(1, -246.22,r'T$\approx390$')

plt.xticks(np.arange(0, 20, 2))
plt.tick_params(axis='both', labelsize = 15)
plt.rc('axes', linewidth=3)



ax=plt.gca()
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
kw = dict(xycoords='data',textcoords="axes fraction", arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
ax.annotate(u'энергия исходной структуры', xy=(8, -246.13), xytext=(0.94,0.96), **kw)


plt.tight_layout()
#plt.grid(True)
plt.savefig('energy_structure.png', dpi = 600)

plt.show()






#plt.plot(a1, b1, 'r--',a2, b2, 'r--',a3, b3, 'r--', linewidth=1.0)
```

```python
def adjustFigAspect(fig,aspect=1):
    '''
    Adjust the subplot parameters so that the figure has the correct
    aspect ratio.
    '''
    xsize,ysize = fig.get_size_inches()
    minsize = min(xsize,ysize)
    xlim = .4*minsize/xsize
    ylim = .4*minsize/ysize
    if aspect < 1:
        xlim *= aspect
    else:
        ylim /= aspect
    fig.subplots_adjust(left=.5-xlim,
                        right=.5+xlim,
                        bottom=.5-ylim,
                        top=.5+ylim)


fig = plt.figure()
fig.clear()
adjustFigAspect(fig,aspect=5)

plt.scatter(calculation['\xb9_structure'],calculation['energy'], c='black')


plt.xlabel(u'Номер структуры', fontsize=12)
plt.ylabel(u'Энергия эл. ячейки, эВ', fontsize=12)

A = 0, 19
B = -246.13, -246.13
plt.plot(A,B, 'r--', linewidth=3.0)
#plt.text(1, -246.22,r'T$\approx390$')

plt.xticks(np.arange(0, 20, 1))
plt.tick_params(axis='both', labelsize = 10)
plt.rc('axes', linewidth=1)



ax=plt.gca()
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
kw = dict(xycoords='data',textcoords="axes fraction", arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
ax.annotate(u'энергия исходной структуры', xy=(8, -246.13), xytext=(0.94,0.96), **kw)

x1 = [7,8,9]
y1 = [-245.73,-245.83,-246.02]
plt.scatter(x1,y1, c='blue')

x2 = [4, 6, 10]
y2 = [-246.215,-246.02127,-245.85063]
plt.scatter(x2,y2, c='red')

x3 = [11, 14, 15]
y3 = [-246.17327,-246.18942,-246.16222]
plt.scatter(x3,y3, c='green')


plt.tight_layout()
#plt.grid(True)
plt.savefig('energy_structure_.png', dpi = 600)

plt.show()
```
