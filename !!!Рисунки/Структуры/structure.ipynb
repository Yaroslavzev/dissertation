```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


from matplotlib import rcParams
rcParams.update({'figure.autolayout': True})
```

```python
structure = pd.read_csv('Structure.txt', '\t', names = ['temperature','r-factor','s','occCu2','V', 'Ueq', 'cu21']) # импорт экспериментальных данных
print structure
```

```python
list(structure)
```

```python
fig = plt.figure()

#ax = fig.add_subplot(111)

#ax.set_xlim(-20,200)
#ax.set_ylim(0,1300)



plt.plot(structure['temperature'],structure['r-factor'], linewidth=2.0)


plt.errorbar(structure['temperature'], structure['r-factor'],            
#             yerr=[0.003,0.003,0.01,0.011,0.011],
            fmt='-o',
           linewidth=2.0, ecolor='g', capthick=2)


plt.xlabel(u'Температура, К', fontsize=22)
plt.ylabel(u'R-фактор, %', fontsize=22)
plt.grid(True)
plt.tick_params(axis='both', labelsize = 15)
plt.rc('axes', linewidth=3)

plt.tight_layout()
plt.savefig('structure_r_factor.png', dpi = 600)



plt.show()

```

```python
fig = plt.figure()

#ax = fig.add_subplot(111)

#ax.set_xlim(-20,200)
#ax.set_ylim(0,1300)

plt.plot(structure['temperature'],structure['occCu2'], linewidth=2.0)

plt.errorbar(structure['temperature'], structure['occCu2'],            
            yerr=[0.003,0.003,0.01,0.011,0.011],
            fmt='-o',
           linewidth=2.0, ecolor='g', capthick=2)

plt.xlabel(u'Температура, К', fontsize=22)
plt.ylabel(u'Заселенность Cu21', fontsize=22)


plt.grid(True)
plt.tick_params(axis='both', labelsize = 15)
plt.rc('axes', linewidth=3)

plt.tight_layout()
plt.savefig('structure_occCu2.png', dpi = 600)

plt.show()
```
